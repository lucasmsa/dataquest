### At first, convert numerical columns that are being stored as string values 🐍

```python
print(f"Price column:\n{autos['price'].head()}\n\nOdometer column:\n{autos['odometer'].head()}")
```

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
autos["price"]
```

```python
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","").astype(float)
autos["odometer"]
```

### Rename the odometer column so that it explicitly denotes that the value is in ``km``

```python
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```

```python
def eliminate_outliers(series: pd.Series, series_name: str):
    print(f"Quantity of unique values on the {series_name} series: {series.unique().shape[0]}")
    print(series.describe())
    series_description_indexing = 5 if series_name == 'Odometer' else 100
    print(f"\n{series_description_indexing} highest values and how many times they appear:\n{series.value_counts().sort_index(ascending=False).head(series_description_indexing)}")
    print(f"\n{series_description_indexing} lowest values and how many times they appear:\n{series.value_counts().sort_index(ascending=False).tail(series_description_indexing)}")
    if series_name == 'Odometer':
        print(f"\nFor the odometer series we cannot see any outliers, the kilometrage is normal 🦫")
    else: 
        print("\nAs we can see the price series contains a lot of outliers on both ends of the spectrum.\nFor this analysis we'll consider the values between 65699$ and 300$, because those values start to resemble normality 🤌🏾")
        return (300.0, 65699.0)
        
eliminate_outliers(autos["odometer_km"], "Odometer")
```

```python
(x, y) = eliminate_outliers(autos["price"], "Price")
autos["price"] = autos.loc[autos["price"].between(x, y), "price"]
```

```python
print(f"New top values:\n{autos['price'].value_counts().sort_index(ascending=False).head()}\n\nNew smallest values:\n{autos['price'].value_counts().sort_index(ascending=False).tail()}")
```

### Exploring date series of the dataset. As we can see, the ``ad_created`` column diverges a lot from the ``last_seen `` and ``date_crawled`` columns, notice that the ``ad_created`` was fetched from the website, whereas ``last_seen`` and ``date_crawled`` were generated by the crawler

```python
def generate_series_date_distribution(series: pd.Series, series_name: str):
    print(f"Dates distribution for the {series_name} column {random.choice(['🦘', '🐋', '🦭', '🐎'])}\n")
    print(f"{series.str[:10].value_counts(normalize=True, dropna=False).sort_index(ascending=False)}\n")

generate_series_date_distribution(autos["ad_created"], "Ad created")
generate_series_date_distribution(autos["last_seen"], "Last seen")
generate_series_date_distribution(autos["date_crawled"], "Date crawled")
```
